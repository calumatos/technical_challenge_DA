```python
data2.drop(['specials', 'refill liters', 'refill gas'], axis=1, inplace=True)
```

```python
data2['temp_dif'] = data2['temp_inside'] - data2['temp_outside']
```

```python
data2['fuel_used'] = data2['distance'] * (data2['consume']/100)
```

```python
data2
```

```python
# Current price in Germany

E10 = 1.75
SP98 = 1.88
```

```python
for i, row in data2.iterrows():
    if row['gas_type'] == 'E10':
        data2.at[i, 'trip_cost'] = E10 * row['fuel_used']
    elif row['gas_type'] == 'SP98':
        data2.at[i, 'trip_cost'] = SP98 * row['fuel_used']
    else:
        data2.at[i, 'trip_cost'] = 0
```

```python
# disntance: in km
# consume: liters per 100km 
# speed: time in minutes
# temp: in celsius

data2 
```
