```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('clean_measurements.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
# Distribution of all the features

for column in data.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[column], bins=100, kde=True)  # Adjust bins and other parameters as needed
    plt.title(f'Histogram for {column}')
    plt.xlabel(column)
    plt.ylabel('Number of trips')
    plt.show()
```

```python
# Calculate and display the correlation matrix for numerical variables

numeric_data = data.select_dtypes(include='number')
corr_matrix = numeric_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix for Numerical Variables')
plt.show()
```

```python
# Use of each type of gas by temperature

g = sns.FacetGrid(data, col="gas_type", height=6)
g.map(plt.hist, "temp_outside", bins=30, alpha=0.7)
g.set_axis_labels("Temperature Outside", "Number of Trips")
g.set_titles(col_template="{col_name} Gas Type")
plt.show()
```

```python
# Use of each type of gas by temperature

g = sns.FacetGrid(data, col="gas_type", height=6)
g.map(plt.hist, "temp_dif", bins=30, alpha=0.7)
g.set_axis_labels("Difference of Temperature", "Number of Trips")
g.set_titles(col_template="{col_name} Gas Type")
plt.show()
```

```python
# The graphs show that for long distances the SP98 are more often used

g = sns.FacetGrid(data, col="gas_type", height=6)
g.map(plt.hist, "distance", bins=30, alpha=0.7)
g.set_axis_labels("Distance in km", "Number of Trips")
g.set_titles(col_template="{col_name} Gas Type")
plt.show()
```

```python
# The graphs show that SP98 has the more expensive trip due the longer distances trips. 

g = sns.FacetGrid(data, col="gas_type", height=6)
g.map(plt.hist, "trip_cost", bins=30, alpha=0.7)
g.set_axis_labels("Cost of the trip", "Number of Trips")
g.set_titles(col_template="{col_name} Gas Type")
plt.show()
```

```python
g = sns.FacetGrid(data, col="gas_type", height=6)
g.map(plt.hist, "speed", bins=30, alpha=0.7)
g.set_axis_labels("Speed", "Number of Trips")
g.set_titles(col_template="{col_name} Gas Type")
plt.show()
```

```python
g = sns.FacetGrid(data, col="gas_type", height=6)
g.map(plt.hist, "consume", bins=30, alpha=0.7)
g.set_axis_labels("Consume", "Number of Trips")
g.set_titles(col_template="{col_name} Gas Type")
plt.show()
```

```python
# This graph show that the higher the speed, the higher is the consume of SP98 comparing to the E10  

sns.lineplot(data=data, x="speed", y="consume", hue="gas_type", palette="viridis", err_style=None)
plt.xlabel("Speed")
plt.ylabel("Consumption")
plt.title("Relationship between Speed and Consumption by Gas Type")
```

### Conclusions:

- SP98 is used more often, although it is more expensive.

- SP98 is often used for long journeys, although it consumes more fuel at higher speeds.

- In terms of use in different temperatures, if the difference is too great, it would be better to use the E10 type.
